## Wrnagling - geo_json ready

```python
import pandas as pd
```

```python
data = pd.read_csv('Data/Cafe__restaurant__bistro_seats_2015_map.csv')
```

```python
data['Industry (ANZSIC4) code'].unique()
```

```python
data[data['Industry (ANZSIC4) code']==9003].head(5)
```

```python
# Remove brackets, dots, split one location column into Latitude and Longitude
def extract(data, code):
    df = data[data['Industry (ANZSIC4) code']==code]
    df.loc[:,'Location'] = df.loc[:,'Location'].str.replace('(','')
    df.loc[:,'Location'] = df.loc[:,'Location'].str.replace(')','')
    df[['Latitude', 'Longitude']] = df.loc[:,'Location'].str.split(',', n=1, expand=True)
    df = df.drop(['Location'], axis=1)
    df = df.reset_index(drop=True)
    return df
```

```python
cafe_restaurant = extract(data, 4511)
pubs = extract(data, 4520)
takeaway = extract(data, 4512)
accomodation = extract(data, 4400)
performing_art = extract(data, 9003)
```

```python
pubs
```

```python
cafe_restaurant.to_csv('./Data/cafe_restaurant.csv')
pubs.to_csv('./Data/pubs.csv')
takeaway.to_csv('./Data/takeaway.csv')
accomodation.to_csv('./Data/accomodation.csv')
performing_art.to_csv('./Data/performing_art.csv')
```

## Extract Public Art

```python
data = pd.read_csv('Public_artworks__fountains_and_monuments.csv')
```

```python
data.head(1)
```

```python
def extract2(data, code):
    df = data[data['Asset Type']==code]
    df.loc[:,'Co-ordinates'] = df.loc[:,'Co-ordinates'].str.replace('(','')
    df.loc[:,'Co-ordinates'] = df.loc[:,'Co-ordinates'].str.replace(')','')
    df[['Latitude', 'Longitude']] = df.loc[:,'Co-ordinates'].str.split(',', n=1, expand=True)
    df = df.drop(['Co-ordinates'], axis=1)
    df = df.reset_index(drop=True)
    return df
```

```python
public_art = extract2(data, 'Art')
public_art.to_csv('public_art.csv')
```
